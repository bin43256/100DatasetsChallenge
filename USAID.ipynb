```python
import pandas as pd
```

```python
pd.reset_option('display.max_rows')
```

```python
df = pd.read_json('ComprehensiveDataInventory20241210.json')
df.head()
```

```python
dataset_df = pd.json_normalize(df['dataset'])
dataset_df.head()
```

```python
dataset_df['id'] = dataset_df.index
dataset_df['id'] 
```

```python
dataset_df.info()
```

```python
dataset_df.head()
```

```python
# Explode the 'keyword' column
kw_df = dataset_df['keyword'].explode()

# Combine values with the same index back into a single row
combined_keywords = kw_df.groupby(kw_df.index).apply(lambda x: ", ".join(x))
dataset_df['keyword'] = combined_keywords
dataset_df['keyword']
```

```python
normalized_df = pd.json_normalize(dataset_df['distribution'])
dataset_df.drop(columns=['distribution'], inplace=True)
```

```python
normalized_df
```

```python
transformed_df = pd.DataFrame()

for index, row in normalized_df.head().iterrows():
    # Create a temporary dictionary to hold mediaType: downloadURL pairs for each row
    print(index,row)
    temp_dict = {}
    for col in row.dropna():  # Ignore NaN values
        if isinstance(col, dict):  # Ensure it's a dictionary
            media_type = col.get('mediaType')
            download_url = col.get('downloadURL')
            if media_type and download_url:  # If both keys exist, add to the temp_dict
                temp_dict[media_type] = download_url
    # Append the temp_dict as a new row to the transformed DataFrame
    transformed_df = transformed_df.append(temp_dict, ignore_index=True)

# Display the transformed DataFrame
print(transformed_df)
```

```python
transformed_df.head()
```

```python
join_df = pd.concat([dataset_df, transformed_df], axis=1)
join_df
```

```python
join_df.info()
```

```python
def list2Str(lst):
    if type(lst) is list:
        return ",".join(map(str, lst))
    else:
        return lst
join_df.apply(lambda x: [list2Str(i) for i in x])
```

```python
join_df.to_csv('USAID_DataInventory',index=False)
```
