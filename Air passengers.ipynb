```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Datasets\Transportation - Air Passengers.xlsx')
df.head()
```

```python

from matplotlib.ticker import FuncFormatter

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x = 'Profile Year', data= df, 

def millions(x, pos):
    return f'{x/1e6:.2f}M'

ax.yaxis.set_major_formatter(FuncFormatter(millions))

plt.xlabel('Year')
plt.ylabel('Passenger Count (Millions)')
plt.title('Airport Passenger Counts Over Years')
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter


def millions(x, pos):
    return f'{x/1e6:.2f}M'
    
fig, ax = plt.subplots(figsize=(14, 6))

for column in df.columns[1:]:
    ax.plot(df['Profile Year'], df[column], marker='o', label=column)
    ax.yaxis.set_major_formatter(FuncFormatter(millions))
    ax.grid(linestyle = '-')
    for x, y in zip(df['Profile Year'], df[column]):
        ax.text(x, y, f'{y/1e6:.2f}M', ha='center', va='bottom', fontsize=9)


plt.xlabel('Year')
plt.xticks(ticks=df['Profile Year'])
plt.ylabel('Passenger Count (Millions)')
plt.title('Airport Passenger Counts Over Years')
plt.legend(loc='best')
plt.show()
```
